```python
import seaborn as sns
import matplotlib.pyplot as plt
# import plotnine as p9
# import altair
```

```python
tips = sns.load_dataset("tips")
```

```python
tips
```

```python
sns.scatterplot(data=tips, x="total_bill", y="tip")
```

```python
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="day")
```

```python
sns.lmplot(data=tips, x="total_bill", y="tip", hue="day", fit_reg=False, col="day")
```

```python
sns.lmplot(data=tips, x="total_bill", y="tip", hue="day", fit_reg=False, col="day", row="time")
```
